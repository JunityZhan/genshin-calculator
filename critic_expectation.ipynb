```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import cufflinks as cf
import math
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode,plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()


critic_score = 200 #更改双爆分




critic_rate = np.arange(1, 100, 1)
critic_dam = critic_score-critic_rate*2
expectation = (100+critic_dam)*critic_rate/100.0 + 100*(1 - critic_rate/100)
fig = go.Figure()
fig.add_trace(go.Scatter3d(
    x=critic_rate, y=critic_dam, z=expectation,
    mode='markers',
    marker_color=expectation/((100+critic_score/2)*critic_score/4/100+100*(1 - critic_score/4/100)),
    marker=dict(showscale=True),
    text=critic_dam/critic_rate,
    opacity=0.7
))
fig.update_traces(marker_size=2)
fig.update_layout(title=dict(text="双爆分="+str(critic_score)+" 双爆分=暴击率*2+暴击伤害"),
                  scene = dict( xaxis_title='暴击率',
                    yaxis_title='暴击伤害',
                    zaxis_title='期望'),
                  margin=dict(l=20, r=20, t=39, b=0),
                  )
```
